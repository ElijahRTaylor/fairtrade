```python
#renaming values
cocoa_unclean_df['cocoa_price_good_conditions'].replace({"More than 1,500 CFA": "More than 1,300 CFA","1,300-1,500 CFA": "More than 1,300 CFA"}, inplace=True)
print(cocoa_unclean_df['cocoa_price_good_conditions'].value_counts())

cocoa_unclean_df['amount_invest_plantation'].replace({"less than 10,000 CFA": "Less than 10,000 CFA"}, inplace=True)
print(cocoa_unclean_df['amount_invest_plantation'].value_counts())
```

```python
#renaming values
cocoa_unclean_df['income_change'].replace({"This year it increased": "Increased","This year it decreased": "Decreased","No change": "It stayed the same"}, inplace=True)
print(cocoa_unclean_df['income_change'].value_counts())

cocoa_unclean_df['area_production_change_next_harvest'].replace({"It has Increased": "Increase","It has decreased": "Decrease","Stay the same": "It has not changed"}, inplace=True)
print(cocoa_unclean_df['area_production_change_next_harvest'].value_counts())

cocoa_unclean_df['ppe_use_frequency'].replace({"Yes, always": "Always","No, never": "Never"}, inplace=True)
print(cocoa_unclean_df['ppe_use_frequency'].value_counts())

cocoa_unclean_df['income_portion'].replace({"Most": "More than half"}, inplace=True)
print(cocoa_unclean_df['income_portion'].value_counts())

cocoa_unclean_df['water_access_distance'].replace({"> 30": "It takes more than 30 minutes", "I have access to drinking water in my home": "Always available in my house", "It is always available in my house": "Always available in my house", "< 30": "It takes less than 30 minutes", "No access": "I don't have access to potable/drinking water"}, inplace=True)
print(cocoa_unclean_df['water_access_distance'].value_counts())

cocoa_unclean_df['skipmeals_frequency'].replace({"Every 6 months": "At least once every six months", "Every 1-2 months": "At least once every two months", "Every month": "At least once a month", "Every week": "At least once a week"}, inplace=True)
print(cocoa_unclean_df['skipmeals_frequency'].value_counts())

cocoa_unclean_df['children_viability_eng'].replace({"Not sure": "Not Sure"}, inplace=True)
print(cocoa_unclean_df['children_viability_eng'].value_counts())

cocoa_unclean_df['leadership_satisfaction'].replace({"Yes, satisfied": "Somewhat Satisfied", "Yes, very satisfied": "Very satisfied", "No, not satisfied": "Somewhat dissatisfied"}, inplace=True)
print(cocoa_unclean_df['leadership_satisfaction'].value_counts())

cocoa_unclean_df['monthly income'].replace({"Between 0 and 110,000 CFA": "55000", "Between 110,000 and 180,000 CFA": "145000", "Between 180,000 and 350,000 CFA": "265000"}, inplace=True)
print(cocoa_unclean_df['monthly income'].value_counts())
```
