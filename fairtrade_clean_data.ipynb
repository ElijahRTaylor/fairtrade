```python
import numpy as np
import pandas as pd
```

```python
# Load the data
fairtrade_data = "IRL_ClassProjectData_220210.xlsx"
fairtrade_df = pd.read_excel(fairtrade_data)
fairtrade_df.head()
```

```python
# Change column names 
fairtrade_df.rename(columns = {'entity_tenure - Years':'entity_tenure_years', 
                               'entity_tenure - Months':'entity_tenure_months'}, inplace = True)

# Drop columns 
fairtrade_df.drop(['Survey Number','Response ID','ch_account_name','gender_core','age_core',
                   'education_level_core','entity_seasons','work_last_season_tf','num_week_income_core_usd',
                   'num_week_income_core_mxn','weeks_per_year','income_change_core','save_money_tf_core',
                   'financial_resilience_core','unsafe_reason_ms_core','management_complaint_treatment_core',
                   'leadership_feedback_core','leadership_satisfaction_core','Unnamed: 33'], axis=1, inplace=True)

fairtrade_df.head(10)
```

```python
# Identify the null rows
fairtrade_df.isnull().sum()
```

```python
# Drop the null rows 
fairtrade_df.dropna(subset = ['entity_tenure_years','entity_tenure_months','injury_last_year','covid_safety_measures_core','return_next_season_core','pressure_to_work_core','factory_recommend_tf_core'], inplace=True)
fairtrade_df.isnull().sum()
```

```python
# Check the datatypes of our columns 
fairtrade_df.dtypes
```

```python
# Drop the rows with 'refused'
fairtrade_df = fairtrade_df[fairtrade_df["understand_contract"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["workplace_satis_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["trust_leadership_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["living_wage_tf_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["conditions_rate_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["safety_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["accident_last_year"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["injury_last_year"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["covid_safety_measures_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["return_next_season_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["pressure_to_work_core"].str.contains("Refused") == False]
fairtrade_df = fairtrade_df[fairtrade_df["factory_recommend_tf_core"].str.contains("Refused") == False]

fairtrade_df.head(10)
```

```python
# Replace values in factory_recommend_tf_core column
fairtrade_df['factory_recommend_tf_core'] = fairtrade_df['factory_recommend_tf_core'].replace(['10 - Extremely likely', '5 - Neutral'], ['10','5'], inplace=True)
fairtrade_df.head(30)
```
