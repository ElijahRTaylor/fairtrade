```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced

```

```python
columns = ["age_core", "education_level_core", "worker_type_core","covid_safety_measures_core","factory_recommend_tf_core"]
target=['factory_recommend_tf_core']
```

```python
file_path=("Resources/IRL_ClassProjectData_220210.xlsx")
df=pd.read_excel(file_path)
df=df.loc[:, columns].copy()


#Drop rows with missing data

df=df.dropna(axis=0)

#Drop rows that contain  refused or don't know

#Define values being dropped
values=["Refused", "Don't know"]


df=df[df.factory_recommend_tf_core.isin(values)==False]

dict= {"10 - Extremely likely ": "10", "0- Extremely unlikely": "0", "5 - Neutral": "5"}
df=df.replace({"factory_recommend_tf_core": dict})

df.reset_index(inplace=True, drop=True)



df.head(44)

```

# Split the Data into Training and Testing

```python
# Create our features
X=pd.get_dummies(df.drop('factory_recommend_tf_core', axis=1))

# Create our target
y=df['factory_recommend_tf_core'].astype(str)



```

```python
X.describe()
```

```python
y.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   random_state=24,
                                                   stratify=y)

X_train.shape
```

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
clf = BalancedRandomForestClassifier(random_state=24, n_estimators=100)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

```python
# List the features sorted in descending order by feature importance
feature_importances = clf.feature_importances_
features = sorted(zip(X.columns, clf.feature_importances_), key = lambda x: x[1], reverse=True)
features
```
